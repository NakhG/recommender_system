```python
user_sim_mat = metrics.pairwise.cosine_similarity(nonan_test)
```

```python
def getUser(user_sim_mat,u):
    score=[]
    for others in range(len(user_sim_mat)):
        if others != u:
            score.append([user_sim_mat[u][others], others])
    score.sort()
    score.reverse()
    return score[:1]
```

```python
getUser(user_sim_mat,0)
```

```python
def standEstItem(nonan_test, u, user_sim_mat, item):
    n = np.shape(nonan_test)[1]
    simTotal = 0.0
    ratSimTotal = 0.0
    
    for j in range(n):
        userRating = nonan_test[u,j]
        if userRating == 0.0:
            continue
        
        sim = user_sim_mat[item][j]
        if sim<=0: sim = 0.0
            
        simTotal += 0
        ratSimTotal += sim * userRating
    
    if simTotal == 0:
        return 0
    else:
        return ratSimTotal/simTotal
        
```

```python
def getItem(nonan_test, user_sim_mat, u):
    #unratedItems = np.nonzero(nonan_test[u])
    
    unratedItems = np.nonzero(np.matrix(nonan_test[u,:]).A == 0)[1]
    
    itemScores = []
    
    for item in unratedItems:
        estimatedScore = standEstItem(nonan_test, u, user_sim_mat, item)
        print(estimatedScore)
        itemScores.append((item,estimatedScore))
        
    return sorted(itemScores, key=lambda jj:jj[1], reverse=True)[:3]
    #for item in unratedItems:
    #    estimatedscore = 
```
